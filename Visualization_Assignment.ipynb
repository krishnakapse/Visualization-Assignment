Q2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.api as sm # Importing statsmodels for qqplot

# Generate dataset of random numbers
np.random.seed(0)
data = np.random.normal(0, 1, 1000)

# Create DataFrame
df = pd.DataFrame({'Values': data})

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['Values'], bins=30, kde=True)
plt.title('Distribution of Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(df['Values'])
plt.title('Boxplot of Values')
plt.show()

# Q-Q Plot
# Using statsmodels.api.qqplot instead of sns.displot
plt.figure(figsize=(8, 6))
sm.qqplot(df['Values'], line='s')  # 'line' argument specifies the reference line type
plt.title('Q-Q Plot of Values')
plt.show()

# Density Plot
plt.figure(figsize=(8, 6))
sns.kdeplot(df['Values'], shade=True)
plt.title('Density Plot of Values')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

Q4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate a dataset with categories and numerical values
data = {
	'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
	'Value': [1, 2, 3, 4, 5, 6, 7, 8, 9]
}

df = pd.DataFrame(data)

# Visualize the distribution of a numerical variable across different categories
plt.figure(figsize=(10,6))

# Group data by category and create a list of values for each category
data_by_category = [df['Value'][df['Category'] == category] for category in df['Category'].unique()]

# labels should be the unique categories
labels = df['Category'].unique() # ['A', 'B', 'C']

# Create the boxplot with the grouped data and labels
plt.boxplot(data_by_category, labels=labels)

# Add title and labels
plt.title('Distribution of Value across Categories')
plt.xlabel('Category')
plt.ylabel('Value')

# Show the plot
plt.show()
```

Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Generate a synthetic dataset with correlated features
np.random.seed(0)
data = np.random.multivariate_normal(mean=[0, 0], cov=[[1, 0.5], [0.5, 1]], size=1000)
df = pd.DataFrame(data, columns=['Feature1', 'Feature2'])

# Visualize the correlation matrix of a dataset using a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', square=True)
plt.title('Correlation Matrix of Dataset')
plt.show()

```

Q3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
np.random.seed(20) data = { 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100), 'Day': np.random.choice(range(1, 31), 100), 'Sales': np.random.randint(1000, 5000, 100) } df = pd.DataFrame(data)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Set the random seed for reproducibility
np.random.seed(20)

# Generate the sales data
data = {
	'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
	'Day': np.random.choice(range(1, 31), 100),
	'Sales': np.random.randint(1000, 5000, 100)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Pivot the data to create a matrix for the heatmap
pivot_df = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')

# Generate a heatmap
plt.figure(figsize=(10,8))
sns.heatmap(pivot_df, cmap='coolwarm', annot=True, fmt='.0f')

# Add title and labels
plt.title('Heatmap of Sales by Month and Day')
plt.xlabel('Day')
plt.ylabel('Month')

# Show the plot
plt.show()
```

Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.embed import components
from bokeh.models import ColumnDataSource, HoverTool

# Generate random x and y values
x = np.random.rand(100)
y = np.random.rand(100)

# Generate random sizes and colors
sizes = np.random.randint(10, 50, 100)
colors = np.random.choice(['red', 'green', 'blue'], 100)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a figure with a title and axis labels
p = figure(title="Scatter Plot", x_axis_label='x', y_axis_label='y')

# Add a scatter plot with different sizes and colors
p.circle(x='x', y='y', size='sizes', color='colors', source=source)

# Add a HoverTool
hover = HoverTool(tooltips=[("x", "@x"), ("y", "@y"), ("size", "@sizes"), ("color", "@colors")])
p.add_tools(hover)

# Show the plot
show(p)
```

Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears'] counts = [20, 25, 30, 35]

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a figure with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count')

# Add a bar chart
p.vbar(x='fruits', top='counts', source=source, width=0.9)

# Show the plot
show(p)
```

Q4. Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30)

```python
import numpy as np
from bokeh.plotting import figure, show

# Generate random data
np.random.seed(0)
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create Bokeh plot
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency',
           x_range=[edges[0], edges[-1]])

# Add histogram renderer to plot
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="#CAB2D6", line_color="white")

# Show plot
show(p)

```

Q5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10) x = np.linspace(0, 1, 10) y = np.linspace(0, 1, 10) xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Generate random data
data_heatmap = np.random.rand(10, 10)

# Create a meshgrid
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(image=[data_heatmap]))

# Create a figure with a title and axis labels
p = figure(title="Heatmap", x_axis_label='x', y_axis_label='y')

# Add an image glyph to create the heatmap
p.image(image='image', x=0, y=0, dw=1, dh=1, source=source)

# Show the plot
show(p)

```
